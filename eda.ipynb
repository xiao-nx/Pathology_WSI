```python
# -*-coding: utf-8 -*-
import os
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker
from matplotlib.pyplot import MultipleLocator

%matplotlib inline
```

```python
data_path = './data/'
fname = os.path.join(data_path, 'minnesota3.csv')
df = pd.read_csv(fname)
df.head()
```

```python
plt.figure(figsize=(20, 6))
confirmed = plt.bar(df['date'], df['cum_confirmed'], color='red', label='Cumulative confirmed cases')
recovered = plt.plot(df['date'], df['recovered'], color='green', label='Cumulative recovered cases')
death = plt.bar(df['date'], df['death'], color='black', label='Cumulative death cases')
#设置刻度的字号
plt.xlabel('Date',fontsize=14)
# 标签旋转
plt.xticks(rotation=45)
#设置x轴标签及其字号
plt.ylabel('Cumulative Cases',fontsize=14)
plt.legend(loc='best')
plt.title('Cumulative confirmed cases in minnesota, the US', fontsize=24)
#设置图大小
# plt.style.use({'figure.figsize':(10, 6)})
#设置坐标轴间隔
x_major_locator= MultipleLocator(10)#把x轴的刻度间隔设置为1，并存在变量里
y_major_locator= MultipleLocator(20000)#把y轴的刻度间隔设置为5，并存在变量里
ax= plt.gca()#ax为两条坐标轴的实例
ax.xaxis.set_major_locator(x_major_locator)#把x轴的主刻度设置为1的倍数
ax.yaxis.set_major_locator(y_major_locator)#把y轴的主刻度设置为5的倍数

# plt.savefig('./figures/total_confirmed.png')
plt.show(block=False)
```
